# Calculating Mean, Mode and Median Of Numerical Columns

```python
# Numerical Columns as List for Calculating Mean, Mode and Median.
num_cols=df.select_dtypes(include=['float64', 'int64']).columns

num_cols
```

```python
statistics = pd.DataFrame({
    'Mean': df[num_cols].mean(),
    'Median': df[num_cols].median(),
    'Mode': df[num_cols].mode().iloc[0],
    'Standard Deviation': df[num_cols].std()
})

# Printing the Basic statistics
statistics
```

# Visualization Part

## Histogram for Each Numerical Column

```python
import matplotlib.pyplot as plt

for column in num_cols:
    df[column].plot(kind='hist',title=f'{column} Distribution',bins=20)
    plt.show()
```

## Boxplot for Each Numerical Column

```python
import matplotlib.pyplot as plt

for column in num_cols:
    df[column].plot(kind='box',title=f'{column} Boxplot')
    plt.show()
```

## Bar Chart Analysis for Categorical Column

```python
#Identifing Categorical columns
cat_cols=df.select_dtypes(include=['object']).columns
cat_cols
```

```python
for column in cat_cols:
    df[column].value_counts().plot(kind='bar', title=f'{column} Frequency')
    plt.show()
```

# Standardization of Numerical Variables

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
stdized_df = df.copy()
stdized_df[num_cols] = scaler.fit_transform(df[num_cols])

# Comparison Before and After Comparisons
comparison = pd.DataFrame({
    'Original': df[num_cols[0]],
    'Standardized': stdized_df[num_cols[0]]
})
comparison
```

```python
comparison.head()
```

# Conversion of Categorical Data into Dummy Variables

```python
# Application of One-hot encoding on categorical columns
df_encode=pd.get_dummies(df,columns=cat_cols,drop_first=True)

df_encode
```
