#### 1.Standard Scaling :
Standard scaling is a data preprocessing technique used to normalize data by transforming its distribution so that it has a mean of 0 and a standard deviation of 1. This is often applied in machine learning to ensure that features with different scales (like age and income) contribute equally to the model, preventing features with larger ranges from dominating the learning process.

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
```

```python
numcols=['age','fnlwgt','education_num','capital_gain','capital_loss','hours_per_week']
```

```python
# Standard Scaling.
stdscaler=StandardScaler()
scale_stddata=stdscaler.fit_transform(data[numcols])
scale_stddata_df=pd.DataFrame(scale_stddata,columns=numcols)
```

```python
scale_stddata_df.head(10)
```

#### 2.MinMax Scaling :
MinMax scaling, also known as normalization, is a technique that rescales data to a fixed range, usually between 0 and 1. This method is especially useful when the data does not follow a Gaussian distribution or when the values have a known minimum and maximum.

```python
# Min-Max Scaling.
minmaxscaler=MinMaxScaler()
scale_minmaxdata=minmaxscaler.fit_transform(data[numcols])
scale_minmaxdata_df=pd.DataFrame(scale_minmaxdata,columns=numcols)
```

```python
scale_minmaxdata_df.head(10)
```

#### 1.Apply One-Hot Encoding to categorical variables with less than 5 categories.

```python
from sklearn.preprocessing import OneHotEncoder,LabelEncoder
```

```python
catcols=data.select_dtypes(include=['object']).columns
unique_catcount={col:data[col].nunique() for col in catcols}
unique_catcount
```

```python
# One-Hot Encoding.
onehotcols=['sex','income']
onehotencoder=OneHotEncoder(sparse_output=False,drop='first')
onehot_data=pd.DataFrame(onehotencoder.fit_transform(data[onehotcols]),columns=onehotencoder.get_feature_names_out(onehotcols))
onehot_data.head(10)
```

#### 2.Use Label Encoding for categorical variables with more than 5 categories

```python
# Label Encoding.
labelcols=['workclass','education','marital_status','occupation','relationship','race','native_country']
labelencoder={col:LabelEncoder() for col in labelcols}
labelencode_data=data[labelcols].apply(lambda col:labelencoder[col.name].fit_transform(col))
labelencode_data.head(10)
```

```python
# Combining the encoded columns with original data.
encode_data=data.drop(columns=onehotcols+labelcols)
encode_data=pd.concat([encode_data,onehot_data,labelencode_data],axis=1)
encode_data.head(10)
```

#### 3.Discuss the pros and cons of One-Hot Encoding and Label Encoding.

##### One-Hot Encoding :
###### Pros :
1. Prevents model assumptions about ordinal relationships (e.g., in categorical variables like colors).
2. Works well for nominal categories (no inherent order).

###### Cons :
1. Increases dimensionality, especially with high-cardinality features.
2. Can result in sparse matrices, increasing memory usage.

##### Label Encoding :
###### Pros :
1. Efficient in terms of memory and computational resources.
2. Suitable for ordinal categories (where order matters).
###### Cons :
1. Imposes an ordinal relationship where there may be none.
2. Can lead to poor performance if used for nominal categories.

#### 2.Apply the PPS (Predictive Power Score) to find and discuss the relationships between features. Compare its findings with the correlation matrix

#### PPS-Predictive Power Score :
The PPS is an asymmetric, data-type-agnostic score that can detect linear or non-linear relationships between two columns. The score ranges from 0 (no predictive power) to 1 (perfect predictive power). It can be used as an alternative to the correlation (matrix).

```python
pip install ppscore
```

```python
import ppscore as pps
pps_matrix=pps.matrix(data)[['x','y','ppscore']].pivot(columns='x',index='y',values='ppscore')
correlation_matrix=data.corr()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(pps_matrix,annot=True,cmap='coolwarm')
plt.title("PPS Matrix")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```
