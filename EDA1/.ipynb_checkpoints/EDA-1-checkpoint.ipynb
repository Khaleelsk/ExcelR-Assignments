#### 4.Detect and treat outliers if necessary.

```python
from scipy import stats # Calculating the Z-score for each column and filtering the out outliers.
z_scores=np.abs(stats.zscore(data))
data=data[(z_scores<3).all(axis=1)]
```

```python
# Using IQR(Interquantile Range) for filtering outliers.
Q1=data.quantile(0.25) # Q1 25th percentile.
Q3=data.quantile(0.75) # Q3 75th percentile.
IQR=Q3-Q1 # Calculating the Interquartile Range(IQR).
data2=data[~((data<(Q1-1.5*IQR))|(data>(Q3+1.5*IQR))).any(axis=1)] # Removing the rows with outliers from the DataFrame.
```

```python
# Comapring the shape of the DataFrame before and after removing outliers
print(data.shape)
print(data2.shape)
```

Implications: Strong correlations may suggest a potential relationship between variables, which could be causal or due to confounding factors. However, correlation does not imply causation.
